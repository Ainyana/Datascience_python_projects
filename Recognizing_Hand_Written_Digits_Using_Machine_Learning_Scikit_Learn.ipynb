```
# https://scikit-learn.org/stable/auto_examples/classification/plot_digits_classification.html#sphx-glr-auto-examples-classification-plot-digits-classification-py

import matplotlib.pyplot as plt
from sklearn import datasets, svm, metrics
from sklearn.model_selection import train_test_split
```

```
import pandas as pd
import numpy as np
```

```
dataset = datasets.load_digits()
print(dataset)
```

```
_, axes = plt.subplots(2,4)
```

```
images_and_labels = list(zip(dataset.images,dataset.target))
```

```
for ax, (image, label) in zip(axes[0, :], images_and_labels[:4]):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    ax.set_title('Training: %i' % label)
```

```
n_samples = len(dataset.images)
data = dataset.images.reshape((n_samples, -1))
```

```
classifier = svm.SVC(gamma=0.001)
classifier
```

```
# split data ito train and test 

X_train, X_test, y_train, y_test = train_test_split(
    data, dataset.target, test_size=0.5, shuffle=False)
```

```
len(X_train)
```

```
classifier.fit(X_train, y_train)
```

```
predicted = classifier.predict(X_test)
```

```
predicted
```

```

images_and_predictions = list(zip(dataset.images[n_samples // 2:], predicted))

images_and_predictions
```

```
for ax, (image, prediction) in zip(axes[1, :], images_and_predictions[:4]):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    ax.set_title('Prediction: %i' % prediction)
```

```
print("Classification report for classifier %s:\n%s\n"
      % (classifier, metrics.classification_report(y_test, predicted)))
```

```
disp = metrics.plot_confusion_matrix(classifier, X_test, y_test)
disp
```

```
disp = metrics.plot_confusion_matrix(classifier, X_test, y_test)
disp.figure_.suptitle("Confusion Matrix")
print("Confusion matrix:\n%s" % disp.confusion_matrix)

plt.show()
```
